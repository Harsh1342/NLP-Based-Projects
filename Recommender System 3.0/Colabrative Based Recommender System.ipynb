## Popularity Based Recommender System

```python
num_rating = df5.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating = num_rating.rename(columns={'Book-Rating':'Number_Of_Ratings'})
num_rating
```

```python
avg_num_rating = df5.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_num_rating = avg_num_rating.rename(columns={'Book-Rating':'avg_Number_Of_Ratings'})
avg_num_rating
```

```python
popularity_df = num_rating.merge(avg_num_rating,on='Book-Title')
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['Number_Of_Ratings']>=250]
popularity_df
```

```python
popularity_df=popularity_df.sort_values('avg_Number_Of_Ratings',ascending=False)
popularity_df = popularity_df.head(50)
popularity_df
```

```python
df6 = popularity_df.merge(df2,on='Book-Title').drop_duplicates('Book-Title')
df6
```

```python
df6.shape
```

```python
df6 = df6[['Book-Title','Book-Author','Number_Of_Ratings','Image-URL-M','avg_Number_Of_Ratings']]
df6
```

## Colabrative Filtering

```python
df4.head()
```

```python
x = df4.groupby('User-ID').count()['Book-Rating']>200
pro_reader = x[x].index
pro_reader
```

```python
filtered_rating = df4[df4['User-ID'].isin(pro_reader)]
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
```

```python

z = filtered_rating['Book-Title'].isin(famous_books)
z[z].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
final_ratings.duplicated().sum()
```

```python
pt= final_ratings.pivot_table(index = 'Book-Title' , columns = 'User-ID' , values = 'Book-Rating')
```

```python
print(pt.shape)
pt
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_score = cosine_similarity(pt)
sim_score
```

```python
sim_score.shape # book-book cos similarity
```

```python
def recommend(book_name):
    
    idx = np.where(pt.index==book_name)[0][0]
    
    list1 = sorted(list(enumerate(sim_score[idx])),reverse = True,key = lambda x: x[1])[1:6]
    
    for i in list1:
        print(pt.index[i[0]])
    
```

```python
dum = np.where(pt.index=='1984')[0][0] ##find index of book
dum
```

```python
lst = sorted(list(enumerate(sim_score[dum])),reverse = True,key = lambda x: x[1])[1:6] ##enumerate = index ,
                                                                            ##key = taking second argument for sorting
```

```python

for i in lst:
        print(i[0] , pt.index[i[0]])
```
